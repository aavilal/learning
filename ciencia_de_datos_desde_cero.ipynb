Asignación de valores salida de una función

```python
def suma_producto(x, y):
    return (x + y), (x * y)
sp = suma_producto(2, 3) # sp es -> (5, 6)
s, p = suma_producto(2, 3) # s es -> 5 y p es -> 6
print("sp=", sp)
print("s=", s)
print("p=", p)
```

Intercambio de valores entre variables

```python
x, y = 1, 2 
print(x, y)
y, x = x, y # Intercambio de valores
print(x, y)
```

## Diccionarios

Conjunto de pares, un valor asociado a una clave

```python
dict_vacio = {}
dict_ejemplo = {
    "variable_x": 100,
    "variable_y": 200
}

print(dict_ejemplo["variable_x"])

# No se pueden llamar claves que no existen
try:
    print(dict_ejemplo["variable_z"])
except KeyError:
    print("No existe la variable z")
    
# Validar si existe una clave
# False si no existe
variable_z = "variable_z" in dict_ejemplo
print(variable_z)
print("\n \n ***** Este método es eficiente para diccionarios grandes *****")
```

Método *get()* como alternativa a levantar la *excepción*

```python
variable_x = dict_ejemplo.get("variable_x", None)
variable_z = dict_ejemplo.get("variable_z", 0)

print("Variable x = ", variable_x)
print("Variable z = ", variable_z)
```

Actualiza los valores de una _clave_

```python
dict_ejemplo = {
    "variable_x": 100,
    "variable_y": 200
}

# Asigna la variable x
variable_x = dict_ejemplo.get("variable_x", None)
print("Variable x = ", variable_x)

# Actualiza el par de la clave variable x
dict_ejemplo["variable_x"] = 300
variable_x = dict_ejemplo.get("variable_x", None)
print("Variable x = ", variable_x)


# Nueva clave

# Asigna la variable z que no existe
variable_z = dict_ejemplo.get("variable_z", 0)
print("Variable z = ", variable_z)

# Asigna el par de la clave z , o sea, la agrega al diccionario
dict_ejemplo["variable_z"] = 1000
variable_z = dict_ejemplo.get("variable_z", 0)
print("Variable z = ", variable_z)
```

Convertir un diccionario en datos estructurados

```python
post_dict = {
    "user": "abraham",
    "text": "data_science is awesome",
    "likes": 25,
     "hashtags": ["#datos", "#sql", "#python", "#machinelearning"]
}

# Otiene todas las claves
post_keys = post_dict.keys()
print(post_keys)

# Obtiene todos los valores
post_values = post_dict.values()
print(post_values)

# Obtiene todos los pares
post_items = post_dict.items()
print(post_items)

# Verifica si existe una clave NO PITONICO
print("Verifica clave NO pitónico: ", "abraham" in post_keys)

# Verifica si existe una clave SI PITONICO
print("Verifica clave SI pitónico: ", "abraham" in post_dict)

# Verifica si existe un valor en valores, es lento pero es la UNICA FORMA.
print("Verifica valores lento, pero único: ", 25 in post_values)
```

## Contadores

Un contador convierte una secuencia de valores en un objeto de tipo defaultdict(int) mapeando claves en contadores.

```python
from collections import Counter
c = Counter([0, 1, 2, 0])
print(c)
```

Una instancia Counter tiene n método most_common que se utiliza mucho

```python
c = Counter("Este es un texto de pruea para el counter")
print(c.most_common(3))
```

## Conjuntos

Un set representa una coleccion de elementos.  
Las principales ventajas de usar conjuntos son:  
- Tener elementos únicos.
- Buscar un elemento de una forma muy rápida.

```python
# Set vacío
s = set()
print(s)

# Agregar un elemento
s.add(1)
s.add(2)
s.add(3)
print(s)

# Cantidad de elementos
x = len(s)
print(x)

# Validar la existencia de un elemento
print(2 in s)
```

## Booleanos

Las siguientes expresiones dan False:  
  
- False
- None
- [] (Lista vacía)
- {} (Dict vacío)
- ""
- set()
- 0
- 0.0

```python
# Todos los elementos de un iterable son verdaderos:
x = all([True, 1, {2}])
print(x)

# Alguno de los elementos del iterable es verdadero
y = any([False, 0, [], None, set(), ""])
print(y)
```

## Ordenar

Solo para ordenar listas se pueden utilizar:  
  
- list.sort()
- sorted(list)

```python
# sort: Ordena una lista existente sobre sí misma
x = [4, 1, 2, 3]
x.sort(reverse=True)
print(x)

# sorted: Crea una nueva lista ordenada
x = [4, 1, 2, 3]
y = sorted(x, reverse=True)
print(y)

# sorted puede usar una clave para ordenar y puede ser una funcion
x = [-4, 1, -2, -3]
x = sorted(x, key=abs, reverse=True)
print(x)

# Un poco más avanzado
x = ["texto_x", "texto_xxx", "texto_xx"]
x = sorted(x, key= lambda x: len(x), reverse=True)
print(x)
```

## Compresiones de listas

Lógica general:  
  
lista = [ expresión -> for x in y <if x ...>]
  
Se crea una lista donde se cumple una operación, para una serie de elementos y con posibles condicionales.

```python
numbers = [x for x in range(10)]
print(numbers)

even_numbers = [x for x in numbers if x % 2 == 0]
print(even_numbers)

squares = [x * x for x in range(3)]
print(squares)

even_squares = [x*x for x in even_numbers]
print(even_squares)
```

Conversión de listas a diccionarios o conjuntos:

```python
even_square_dict = {x: x * x for x in even_numbers}
print(even_square_dict)

even_square_set = {x * x for x in even_numbers}
print(even_square_set)
```

Se usa _ si no se utiliza más la variable, solo para ese punto.

```python
zeros = [0 for _ in even_numbers]
print(zeros)
```

Varios for dentro de una compresión de lista:

```python
increasing_pairs = [(x, y)
    for x in range(1)
    for y in range(1)
]
print(increasing_pairs)

# Mueva el valor dentro de range() para entender mejor
```
