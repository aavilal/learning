## Tuplas

La Tupla es muy parecida a una lista

```python
mi_lista = [1, 2]
mi_tupla = (1, 2)
otra_tupla = 3, 4 # -> e traduce en (3, 4)
mi_lista[1] = 3 # Agrega el 3 en la 2da posición
```

Una Tupla no se puede modificar

```python
try:
    mi_tupla[1] = 3
except TypeError:
    print("Una Tupla no se puede modificar")
```

Asignación de valores salida de una función

```python
def suma_producto(x, y):
    return (x + y), (x * y)
sp = suma_producto(2, 3) # sp es -> (5, 6)
s, p = suma_producto(2, 3) # s es -> 5 y p es -> 6
print("sp=", sp)
print("s=", s)
print("p=", p)
```

Intercambio de valores entre variables

```python
x, y = 1, 2 
print(x, y)
y, x = x, y # Intercambio de valores
print(x, y)
```
