## Tuplas

La Tupla es muy parecida a una lista

```python
mi_lista = [1, 2]
mi_tupla = (1, 2)
otra_tupla = 3, 4 # -> e traduce en (3, 4)
mi_lista[1] = 3 # Agrega el 3 en la 2da posición
```

Una Tupla no se puede modificar

```python
try:
    mi_tupla[1] = 3
except TypeError:
    print("Una Tupla no se puede modificar")
```

Asignación de valores salida de una función

```python
def suma_producto(x, y):
    return (x + y), (x * y)
sp = suma_producto(2, 3) # sp es -> (5, 6)
s, p = suma_producto(2, 3) # s es -> 5 y p es -> 6
print("sp=", sp)
print("s=", s)
print("p=", p)
```

Intercambio de valores entre variables

```python
x, y = 1, 2 
print(x, y)
y, x = x, y # Intercambio de valores
print(x, y)
```

## Diccionarios

Conjunto de pares, un valor asociado a una clave

```python
dict_vacio = {}
dict_ejemplo = {
    "variable_x": 100,
    "variable_y": 200
}

print(dict_ejemplo["variable_x"])

# No se pueden llamar claves que no existen
try:
    print(dict_ejemplo["variable_z"])
except KeyError:
    print("No existe la variable z")
    
# Validar si existe una clave
# False si no existe
variable_z = "variable_z" in dict_ejemplo
print(variable_z)
print("\n \n ***** Este método es eficiente para diccionarios grandes *****")
```

Método *get()* como alternativa a levantar la *excepción*

```python
variable_x = dict_ejemplo.get("variable_x", None)
variable_z = dict_ejemplo.get("variable_z", 0)

print("Variable x = ", variable_x)
print("Variable z = ", variable_z)
```

Actualiza los valores de una _clave_

```python
dict_ejemplo = {
    "variable_x": 100,
    "variable_y": 200
}

# Asigna la variable x
variable_x = dict_ejemplo.get("variable_x", None)
print("Variable x = ", variable_x)

# Actualiza el par de la clave variable x
dict_ejemplo["variable_x"] = 300
variable_x = dict_ejemplo.get("variable_x", None)
print("Variable x = ", variable_x)


# Nueva clave

# Asigna la variable z que no existe
variable_z = dict_ejemplo.get("variable_z", 0)
print("Variable z = ", variable_z)

# Asigna el par de la clave z , o sea, la agrega al diccionario
dict_ejemplo["variable_z"] = 1000
variable_z = dict_ejemplo.get("variable_z", 0)
print("Variable z = ", variable_z)
```

Convertir un diccionario en datos estructurados

```python
post_dict = {
    "user": "abraham",
    "text": "data_science is awesome",
    "likes": 25,
     "hashtags": ["#datos", "#sql", "#python", "#machinelearning"]
}

# Otiene todas las claves
post_keys = post_dict.keys()
print(post_keys)

# Obtiene todos los valores
post_values = post_dict.values()
print(post_values)

# Obtiene todos los pares
post_items = post_dict.items()
print(post_items)

# Verifica si existe una clave NO PITONICO
print("Verifica clave NO pitónico: ", "abraham" in post_keys)

# Verifica si existe una clave SI PITONICO
print("Verifica clave SI pitónico: ", "abraham" in post_dict)

# Verifica si existe un valor en valores, es lento pero es la UNICA FORMA.
print("Verifica valores lento, pero único: ", 25 in post_values)
```

## Contadores

O Counters, convierte una secuencia de valores en un objeto de tipo defaultdict(int) mapeando claves en contadores:

```python
from collections import Counter
c = Counter([1, 2, 2, 3, 3, 3])
print(c)
print(type(c))

# Ejemplo con document

document = """
Fragmentos que capturan la magia y el realismo, como "Es mejor mirar al cielo que vivir allí
"""
word_counts = Counter(document)

# Una instancia Counter tiene un Método most_common:
# Que imprime los n elementos más comunes del conteo.

for word, count in word_counts.most_common(3):
    print(word, count)
```

**IMPORTANTE**: Una _instancia_ deriva de una _clase_, la cual tiene asociados n _métodos_.

## Conjuntos _(sets_)

```python
primes_below_10 = {1, 3, 5, 7}

# Conjunto vacío no es {}

s = set() # Conjunto vacío
print("Vacío: ", s)

s.add(1) # Agregar 1
print("Con 1: ", s)

s.add(2) # Agregar 2
print("Con 2: ", s)

s.add(2) # Agrega 2 pero el conjuto sigue siendo {1, 2}
print("Con doble 2: ", s)
print(s)

y = 2 in s
z = 3 in s

print(y, z)
```

Los conjuntos se utilizan princen primer instancia para tener valores únicos, o buscar dentro de grandes listas.  
Para esa tarea, es mejor un **conjunto**.

```python
stopwords_list = ["como", "el", "la", "en"]
print("tal" in stopwords_list) # False pero es lento porque verifica cada elemento de la lista

stopwords_set = set(stopwords_list)
print("tal" in stopwords_set) # Muy rápido de comprobar
```

La segunda razón es encontrar elementos distintos de una colección.

```python
item_list = [1, 2, 2, 1, 0, 5, 3, 4, 2]
num_items = len(item_list)
print("All items: ", (num_items))
item_set = set(item_list)
num_distinct_items = len(item_set)
print("Unique items: ", num_distinct_items)
```

La frecuencia de uso de los conjuntos es:

1.- Listas  
2.- Diccionarios  
3.- Conjuntos
